```python
def DrawMap(map_center, yelp):
    colors=GetColors()
    map = folium.Map(location=map_center, zoom_start=13)
    for b in yelp.businesses:
        coord = (b.location.coordinate.latitude,b.location.coordinate.longitude)
        popup = "%s (%.1f)" % (b.name, b.rating)
        marker = folium.Marker(location=coord, 
                    popup=popup,
                    icon=folium.Icon(color=colors[b.rating]))
        map.add_child(marker)
    return map
```

```python
def GetColors():
    colors = { 
        0: "red",
        0.5: "red",
        1.0: "red",
        1.5: "red",
        2.0: "red",
        2.5: "red",
        3.0: "green",
        3.5: "green",
        4.0: "blue",
        4.5: "blue",
        5.0: "purple",
    } 
    return colors
```
